# Import the Dependencies 

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import r2_score
```

```python
gold_dataset = pd.read_csv("gold_price_china.csv")
```

```python
# Dropping rows with missing values (if any)
gold_dataset = gold_dataset.dropna()
```

# Correlation

```python
# +ive   ,,values are directly proportinal 
# -ive  ,, values are inversaly proportinal 
correlation = gold_dataset.select_dtypes(include=['number']).corr()
print(correlation)
```

# Graph Visualization

```python
plt.figure(figsize=(8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={"size": 8}, cmap="Blues" )
plt.show()
```

# spliting the label and Features

```python
y = gold_dataset["amount"]
x = gold_dataset.drop(["Unnamed: 0", 'trade_date', 'amount'], axis=1)
```

# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling to features (X)
X_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

# Model Training

```python
# Random forest 
rfr_model = RandomForestRegressor(n_estimators=100)
```

```python
rfr_model.fit(x_train, y_train)
```

```python
# Decsion Tree 
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train, y_train)
```

# MOdel Evaluation

```python
# test data prediction 
test_data_pred = rfr_model.predict(x_test)
# print(test_data_pred)
test_error_score = r2_score(test_data_pred, y_test)
# display value 
print("R square Error of Random Forest : ", test_error_score*100)
# print("Accuracy Random Forest : ", accuracy_score(test_error_score, y_test))
```

```python
rfr_model.score(x_test, y_test)*100
```

```python
# test data prediction 
test_data_pred = dt_model.predict(x_test)
# print(test_data_pred)
test_error_score = r2_score(test_data_pred, y_test)
# display value 
print("R square Error of Decision tree : ", test_error_score*100)
# print("Accuracy Random Forest : ", accuracy_score(test_error_score, y_test))
```

# Compare the Actual And Predicted VAlues

```python
# but first of all we should convert the y_test into list (list is pandas series function)
y_test = list(y_test)
```

```python
# Training data visualization 
plt.plot(y_test, color="blue", label="Actual Value")
plt.plot(test_data_pred, color="green", label="Predicted Value")
plt.title("Actual prices VS Predicted prices")

plt.xlabel("Number of Samples")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
