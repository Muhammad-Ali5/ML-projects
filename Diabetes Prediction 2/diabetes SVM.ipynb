```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes = pd.read_csv("Diabetes_prediction.csv")
```

```python
x = diabetes.drop(columns="Diagnosis", axis=1)
y = diabetes["Diagnosis"]
```

```python
print(x)
```

```python
print(y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standarized_data = scaler.transform(x)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = diabetes["Diagnosis"]
```

```python
print(x)
print("---------------------------------------------------------")
print(y)
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1, stratify=y)
```

```python
print(x.shape , x_test.shape)
```

```python
classifier = svm.SVC(kernel="linear")
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_pred = classifier.predict(x_train)
train_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print("Accuracy score for training data : ",train_data_accuracy*100)
```

```python
x_test_pred = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
```

```python
print("Accuracy score for testing data : ",test_data_accuracy*100)
```
