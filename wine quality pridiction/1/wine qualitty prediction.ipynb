```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

# Correlation

```python
# positive correlation
# negative corr
correlation = wine_dataset.corr()
plt.figure(figsize=(10, 12))
sns.heatmap(correlation, annot=True, cmap="coolwarm", cbar=True, square=True, fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Wine Dataset")
plt.show()

```

```python
# splitting the features
x = wine_dataset.iloc[:, :-1]
y = wine_dataset["quality"]
```

```python
y
```

## label Benarization

```python
y = wine_dataset["quality"].apply(lambda y_value: 1 if y_value >= 7 else 0 )
```

# Trian Test splitting 

```python
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(y.shape, y_train.shape, x_train.shape)
```

# train the model 

```python
r_model = RandomForestClassifier()
```

```python
r_model.fit(x_train, y_train)
```

# MOdel evaluation

```python
# training data accuracy

x_test_pred = r_model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred, y_test)
```

```python
print("Accuracy are : ", test_data_accuracy*100)
```

# Save and Load Model

```python
import joblib
```

```python
#  save with the name 
model_name = "Radom_f_t wine qualitty prediction.pkl"

joblib.dump(r_model, model_name)
```

```python
# now we load our model 
loaded_model = joblib.load(model_name)

print(loaded_model.predict(x_test))
```
