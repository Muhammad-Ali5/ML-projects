```python
# splitting the features 
x = wine_data.iloc[:, :-1]
y = wine_data["quality"]
```

```python
# Label Binarization?
# Label binarization is the process of converting categorical labels into binary values (0 and 1) for classification tasks.
# Extract the 'quality' column
quality_values = wine_data["quality"]
# Define a function to classify wine as good (1) or not good (0)
def classify_quality(y_value): 
    if y_value>=9: 
        return 1
    else :
        return 0

# Apply the function to each quality value
y = quality_values.apply(classify_quality)
```

```python
y
```

```python
wine_data["quality"].unique()
```

```python
#  train test splitting 
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
# train the model 
r_model = RandomForestClassifier()

r_model.fit(x_train, y_train)
```

```python
x_test_pred = r_model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred, y_test)
```

```python
print("Accuracy : ", test_data_accuracy*100)
```
